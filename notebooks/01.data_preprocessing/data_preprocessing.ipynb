### load csv dataset

```python
data = pd.read_csv('../../data/raw/mozambique_rede-escolar_wis_school-data.csv')
```

### check for null values and duplicates

```python
# dataset shape
print(f"rows: {data.shape[0]}")
print(f"columns: {data.shape[1]}")
print()

# check for null values
print('null values:')
print(data.isnull().sum())
print()

# check for duplicates
duplicated = data[data.duplicated()]
print("Duplicates:")
print(duplicated)

# drop both
data.dropna(inplace=True)
data = data.drop_duplicates()
```

### drop columns that will not be used or contain inconsistent data

```python
data.drop([
    'Posto',
    'Localidade',
    'Povoado',
    'ZIP',
    'Cod__Escola',
    'Latrina',
    'TotalLatri',
    'Estado_de_Conservação3',
    'Urinois',
    'TotalUrino',
    'UrinoisFun', 
    'Estado_de_Conservação4'], axis=1, inplace=True)
```

### rename remain columns

```python
data = data.rename(columns={
    'NomeEsco_1':'NomeEscola',
    'QualiAgua':'QualidadeAgua',
    'TipoEnergi':'TipoEnergia',
    'CBFunci':'CasaBanhoFuncional',
    'TotalCB': 'TotalCasaBanho'})
```

### handle some columns inconsistency

```python
data.loc[data['Agua'] == 'Não', 'TipoFonte'] = 'Nenhuma Fonte'
data.loc[data['TipoFonte'] == 'Nenhuma Fonte', 'Agua'] = 'Não'
data.loc[data['Energia'] == 'Não', 'TipoEnergia'] = 'Nenhum'
data.loc[data['TipoEnergia'] == 'Nenhum', 'Energia'] = 'Não'
data.loc[data['CasaBanho'] == 'Não', 'CasaBanhoFuncional'] = 'Não'
```
